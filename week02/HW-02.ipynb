```python
import random 
import matplotlib.pyplot as plt
import numpy as np

trials = 500

def fair_die_simulation(trials):
    """
    returns an array of randomly numbers between 1 and 6. 
    """
    options = [1,2,3,4,5,6]
    simulations = [random.randint(1,6) for i in range(trials)]
    return simulations
```

```python
# List of random fair simulations

fair_list = fair_die_simulation(trials)
print("Fair List = ",fair_list)
plt.hist(fair_list)
plt.xlabel("faces on a die")
plt.ylabel("probabilities")
plt.title("Fair simulations of a die")
```

```python
# result1 is the list of fair_list functions repeated for 100 times

trial_count = 100
result1 = []
for i in range(trial_count):
    x = fair_die_simulation(trials)
    result1.append(x)
print(result1)
```

```python
# Mean for fair simulations
sum_of_list = 0
fair_mean_list = []

for i in range(trial_count):
    for j in range(trials):
            sum_of_list = sum_of_list + result1[i][j]
    mean = sum_of_list/trials
    sum_of_list = 0
    fair_mean_list.append(mean)

print("Fair mean list =",fair_mean_list)
```

```python
# Plot of mean for fair simulations

plt.hist(fair_mean_list)
plt.xlabel("mean")
plt.title("Mean for fair simulation")
```
