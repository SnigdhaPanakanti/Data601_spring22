# Week 11 Homework - Getting data


1. Load the dataframe saved from this week's lab.
1. Scrape data from this page https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_GDP_(nominal)
1. Append to the dataframe: projected GDP (nominal) from IMF estimates from 2022 to 2027. (If data is not available for the country use None)
1. Add a new column and calculate average increase in terms of percentage. 
1. Sort by this column in descending order.

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
df1 = pd.read_csv("df.csv")

df1
```

```python
response = requests.get('https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_GDP_(nominal)')
response.ok
```

```python
response.text[:1000]
```

```python
html_doc = response.text
```

```python
df = pd.read_html(html_doc)[5]
df2 = pd.DataFrame(df)
df2
```

```python
df2.drop(columns = ['2020','2021'],inplace=True)
df2.rename(columns = {'Country (or dependent territory)':'Country_Name'},inplace=True)
df2
```

```python
data = pd.merge(df1, df2,how='outer',on='Country_Name')
data.drop(columns = {'Unnamed: 0'}, inplace = True)
```

```python
data
```

```python
avg_sum = data['2022'] + data['2023'] + data['2024'] + data['2025'] + data['2026'] + data['2027']
```

```python
data['Average(%)'] = (avg_sum/6)*100
```

```python
data
```

```python
d1 = ((data['2023'] - data['2022'])/data['2022'])*100
d2 = ((data['2024'] - data['2023'])/data['2023'])*100
d3 = ((data['2025'] - data['2024'])/data['2024'])*100
d4 = ((data['2026'] - data['2025'])/data['2025'])*100
d5 = ((data['2027'] - data['2026'])/data['2026'])*100

avg_diff = (d1+d2+d3+d4+d5)/5
```

```python
data['Average Increment(%)'] = avg_diff
data
```

```python
data.sort_values(by = 'Average Increment(%)',ascending = False)
```
