## Importing Libraries:

```python
import pandas as pd
import numpy as np
import os
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv(r"C:\Users\Krishna\Desktop\yelp_training_set_business.csv")
df1.shape
```

```python
df1.head()
```

```python
df2 = pd.read_csv(r"C:\Users\Krishna\Desktop\yelp.csv")
df2.shape
```

```python
df2.head()
```

```python
# Merging datasets

df = pd.merge(df1,df2,on='business_id')
df.shape
```

## Data Cleaning:

```python
df.drop(df.filter(regex="Unnamed"),axis=1, inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Dropping Unwanted columns.

df=df.drop(columns=['business_neighborhoods','text','type','cool','useful','funny','business_open','date','business_latitude','business_longitude'])
```

```python
# Replacing NaN values with the most repeated values in the column.

df = df.fillna(df['business_categories'].value_counts().index[0])
```

## Sanity check

```python
df.isnull().sum()
```

```python
df.head()
```

## Hypothesis-1:

- Which city in Arizona has highest number of ratings?

```python
df3 = df.groupby(['stars','business_city']).size().reset_index(name='count')
```

```python
df3 = df3.sort_values(by='count',ascending=False)
```

```python
df3
```

```python
df4 = df3.head(20)
df4
```

```python
df4.groupby(['business_city','stars']).sum().unstack().plot(kind='bar',stacked=True, figsize=(7, 6))
```

- From the above graph, Phoenix city has highest number of ratings in Arizona

## Hypothesis - 2:

- Which city in Arizona has better star rating for restaurants?

```python
flattened_col = pd.DataFrame([(index, value) for (index, values) in df['business_categories'].iteritems() for value in str(values).split(',')],
                             columns=['index', 'business_categories']).set_index('index')
df5 = df.drop('business_categories', axis=1).join(flattened_col)
```

```python
final_df = df5[df5['business_categories'].str.lower().isin(['restaurants'])]
```

```python
final_df
```

```python
restraunt_df = final_df.groupby(['business_city','stars']).size().reset_index(name='Count')
restraunt_df
```

```python
ax = restraunt_df.plot(x="business_city",y="stars",kind="bar")
restraunt_df.plot(x="business_city",y='Count',kind="bar",ax=ax,color="maroon",figsize=(15,8))
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax = plt.subplot()
ax.bar(restraunt_df["business_city"],restraunt_df["stars"])
ax.bar(restraunt_df["business_city"],restraunt_df["Count"],color="green")
```

- Phoenix has better ratings for restaurants compared to other cities in Arizona.

## Hypothesis - 3:

- Which restaurants has high review counts with respect to star ratings?

```python
business_df = final_df[['business_name','stars','business_review_count']]
business_df
```

```python
business_df.groupby(['business_name','stars']).sum().sort_values(by='stars',ascending=False)
```

```python
temp_df1 = business_df.loc[business_df['stars'] == 5]
temp_df2 = business_df.loc[business_df['stars'] == 4]
temp_df3 = business_df.loc[business_df['stars'] == 3]
temp_df4 = business_df.loc[business_df['stars'] == 2]
temp_df5 = business_df.loc[business_df['stars'] == 1]
```

```python
temp_df1.groupby(['business_name']).sum().plot(kind='pie', y='business_review_count',figsize=(10,10),autopct='%1.0f%%')
temp_df2.groupby(['business_name']).sum().plot(kind='pie', y='business_review_count',figsize=(10,10),autopct='%1.0f%%')
temp_df3.groupby(['business_name']).sum().plot(kind='pie', y='business_review_count',figsize=(10,10),autopct='%1.0f%%')
temp_df4.groupby(['business_name']).sum().plot(kind='pie', y='business_review_count',figsize=(10,10),autopct='%1.0f%%')
temp_df5.groupby(['business_name']).sum().plot(kind='pie', y='business_review_count',figsize=(10,10),autopct='%1.0f%%')
```

From the above pie charts, we can see the review counts for each star rating of each restaurants.
