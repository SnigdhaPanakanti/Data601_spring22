```python
import seaborn as sns
import pandas as pd

diamonds = sns.load_dataset("diamonds")

diamonds.head()
```

#### 1.

- Plot the scatter plot of price and carat. Add clarity into the plot as hue. 

- Do you see any pattern? Is it parallel to your expectation?

```python
# Scatter plot of carat vs price.

sns.scatterplot(data = diamonds, x="carat", y="price")
```

```python
# Scatter plot of carat vs price with 'clarity' as hue.

x = diamonds.carat
y = diamonds.price
h = diamonds.clarity.values 
plot = sns.scatterplot(x= x, y = y, hue = h, alpha = 0.1 )
```

#### 2.

- Consider `clarity` and `color` is there any pattern in their heatmap?

```python
plot = sns.displot(diamonds, x="clarity", y="color", cbar = True)
plot.ax.tick_params(labelrotation = 90)
```

#### 3.

- Use `carat` variable as as the size of the points in scatter plot in which x-axis is the `depth` and `price` on the y-axis.

```python
sns.relplot(data = diamonds, x = "depth", y = "price", size = "carat", sizes = (20,200), alpha = .5, height = 6)
```
