```python
import pandas as pd
import re
```

```python
spicy = 'https://raw.githubusercontent.com/msaricaumbc/DS_Data/master/strings/hw_data_spicy.csv'
less_spicy = 'https://raw.githubusercontent.com/msaricaumbc/DS_Data/master/strings/hw_data_less_spicy.csv'
```

```python
df = pd.read_csv(less_spicy, index_col=0)
```

```python
df.head()
```

```python
start_date=[]
end_date=[]
phone_number = []
```

```python
for _,i in df["notes"].iteritems():
    search = i.split("end") if 'end' in i else i.split("ed")
    print(search)
    sd = search[0].replace("start","")
    start_date.append(sd)
    print(search[1])
    con=re.compile('[Pp](#|\/n|n)(:)?()+') 
    if(con.search(search[1])):
        mo=con.search(search[1])
        print(mo.group())
        e=search[1].split(mo.group())
        print(e)
        end_date.append(e[0])
        phone_number.append(e[1])
    else:
        end_date.append(search[1])
        phone_number.append(None)
```

```python
print(*start_date,sep="\n")
```

```python
print(*end_date, sep="\n")

pattern = re.compile(r'\d{4}-\d{2}-\d{2}')
mo = pattern.search('*end_date')
mo
```
