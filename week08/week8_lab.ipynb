- In  `Q12_Part_4_TEXT` they ask the tools the participants are using day to day. However as you might observe the answers are not standardized. 

- Find all the answers with 'jupyter' mentioned in it.

- Find all the participant who are Python users.

- Find all R users. 

- Find all the answers with 'jupyter' mentioned in it.

```python
df.fillna('',inplace = True)
```

```python
df[df['Q12_Part_4_TEXT'].str.lower().str.contains('jupyter')]['Q12_Part_4_TEXT']
```

- Find all the participant who are Python users.

```python
df[df['Q12_Part_4_TEXT'].str.lower().str.contains('pyth|^py')]['Q12_Part_4_TEXT']
```

```python
df[df['Q12_Part_4_TEXT'].str.lower().str.contains(r'^py')]['Q12_Part_4_TEXT']
```

```python
for _, i in df[df['Q12_Part_4_TEXT'].str.lower().str.contains('pyth')]['Q12_Part_4_TEXT'].iteritems():
    print(i)
```

- Find all R users.

```python
df[df['Q12_Part_4_TEXT'].str.lower().str.contains('^r[s ]',regex = True)]['Q12_Part_4_TEXT']
```

```python
df[df['Q12_Part_4_TEXT'].str.lower().str.contains('^r[s ]|^r|^R|',regex = True)]['Q12_Part_4_TEXT']
```

```python
for _,i in df[~df['Q12_Part_4_TEXT'].str.lower().str.contains('^r[s ]|^r|(r)',regex = True)]['Q12_Part_4_TEXT'].iteritems():
    print(i)
```

```python
df[df['Q12_Part_4_TEXT'].str.lower().str.contains('^r[ ,]|[ ,r]r$|r\s?st|rscri|^r$|r\w{2,3}dio',regex = True)]['Q12_Part_4_TEXT']
```

```python
for _,i in df[~df['Q12_Part_4_TEXT'].str.lower().str.contains('^r[ ,]|[ ,r]r$|r\s?st|rscri|^r$|r\w{2,3}dio',regex = True)]['Q12_Part_4_TEXT'].iteritems():
    print(i)
```
